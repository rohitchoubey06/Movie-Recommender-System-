### IV Cleaning genres column using function and ast.literal_eval()

```python
movies.genres.iloc[0]
```

```python
movies.dtypes
```

```python
import ast
```

```python
def convert(obj):
    l1 = []
    for i in ast.literal_eval(obj):
        l1.append(i['name'])
    return l1
```

```python
movies.genres = movies.genres.apply(convert)
```

```python
movies.head()
```

### V Doing the same thing for keywords, cast and crew

```python
movies.keywords = movies.keywords.apply(convert)
```

```python
movies.head()
```

```python
# We will extract only first three names from the cast column
movies.cast.iloc[0:3]
```

```python
def convert2(obj):
    l1 = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3: 
            l1.append(i['name'])
            count+=1
        else:
            break
    return l1
```

```python
movies.cast = movies.cast.apply(convert2)
```

```python
movies.head()
```

```python
movies.crew.iloc[0]
```

```python
def concrew(x):
    l2 = []
    for i in ast.literal_eval(x):
        if i['job'] == 'Director':
            l2.append(i['name'])
            break
    return l2
```

```python
movies.crew = movies.crew.apply(concrew)
```

```python
movies.head(3)
```

```python
movies['Director'] = movies['crew']
```

```python
movies = movies.drop('crew',axis=1)
```

### Let's perform stemming

```python
import nltk
```

```python
df.tags.iloc[0]
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return ' '.join(y)
        
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. samworthington zoesaldana sigourneyweaver jamescameron cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d action adventure fantasy sciencefiction')
```

```python
stem('dancing')
```

```python
df.tags= df.tags.apply(stem)
```

```python
# Extracting the Index of movie
df[df['title']=='Batman Begins'].index[0]
```

```python
def recommend(movie):
    movie_index = df[df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(df.iloc[i[0]].title)
    
```
